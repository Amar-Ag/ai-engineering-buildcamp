```python
from pathlib import Path

video_id = 'r0gHbW_MmaQ'
subtitles_file = Path(f'subtitles_{video_id}.txt')
subtitles = subtitles_file.read_text(encoding='utf-8')
```

```python
print(subtitles[:200])
```

```python
lines = subtitles.split('\n')
```

```python
from gitsource import chunk_documents

chunks = chunk_documents(
    [{'content': lines}],
    size=100,
    step=50
)
```

```python
for chunk in chunks:
    chunk['content'] = ' '.join(chunk['content'])
```

```python
from minsearch import Index

index = Index(text_fields=['content'])
index.fit(chunks)
```

```python
from rag import RAG
from google import genai
import os

client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
video_rag = RAG(index=index, llm_client=client)
answer = video_rag.rag("'what is covered in week 3?")
print(answer.answer)
```
